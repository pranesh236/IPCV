```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('messi.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
h, w = img_rgb.shape[:2]
```

```python
center = (w//2, h//2)
angle = 45
scale = 1.0
```

```python
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated = cv2.warpAffine(img_rgb, rotation_matrix, (w, h))
```

```python
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original")
```

```python
plt.subplot(1,2,2)
plt.imshow(rotated)
plt.title("Rotated 45°")
plt.show()
```
