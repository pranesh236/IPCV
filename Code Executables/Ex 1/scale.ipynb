```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('messi.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
s1=3.0
s2=1/3.0
```

```python
h1, w1 = img_rgb.shape[:2]
```

```python
#Zoom
h2 = int(h1*s1)
w2 = int(h2*s1)
```

```python
#Shrink
h3 = int(h1*s2)
w3 = int(w1*s2)
```

```python
zoom=cv2.resize(img_rgb, (h2,w2), interpolation=cv2.INTER_CUBIC)
shrink = cv2.resize(img_rgb, (h3,w3), interpolation=cv2.INTER_AREA)
```

```python
plt.subplot(1,3,1); plt.imshow(img_rgb); plt.title("Original")
```

```python
plt.subplot(1,3,2); plt.imshow(zoom); plt.title("Zoom")
```

```python
plt.subplot(1,3,3); plt.imshow(shrink); plt.title("Shrink")
```
